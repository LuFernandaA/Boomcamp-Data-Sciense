**Parte 2 — Limpieza de datos (Pandas)**📌

```python
df1
```

```python
#Convierto vacios y Estandarizo/Mapeo la columna de pais

df1['pais'] = df1['pais'].str.lower()

df1['pais'] = df1['pais'].replace("",None)
df1['pais'] = df1['pais'].replace(" ",None)

df1['pais'] = df1['pais'].replace(
    {"col":"Colombia",
     "COL":"Colombia",
     "colombia":"Colombia",
     "peru":"Peru",
     "perú":"Peru",
     "chile":"Chile"}
)

#Estandarizo/Mapeo la columna de metodo_pago

df1['metodo_pago'] = df1['metodo_pago'].str.lower()
df1['metodo_pago'] = df1['metodo_pago'].replace({
    "transferencia":"Transferencia",
    "efectivo":"Efectivo",
    "tarjeta":"Tarjeta",
    "TRANFERENCIA":"Transferencia"})


df1['metodo_pago'] = df1['metodo_pago'].replace("",None)
df1['metodo_pago'] = df1['metodo_pago'].replace(" ",None)
```

```python
#Manejo nulos en precio

df1['precio'] = df1['precio'].replace("",None)
df1['precio'] = df1['precio'].replace(" ",None)
df1['precio'] = df1['precio'].fillna(df1['precio'].mean())
```

```python
#Convierto, corrijo y manejo nulos numericos en la columna de cantidad

df1['cantidad'] = pd.to_numeric(df1['cantidad'], errors='coerce')
df1['cantidad'] = df1['cantidad'].fillna(df1['cantidad'].mean())
df1['cantidad'] = df1['cantidad'].round(1)
```

```python
#Convierto columna fecha a formato de fecha

df1['fecha'] = pd.to_datetime(df1['fecha'],errors='coerce')
df1 = df1.dropna(subset=['fecha'])
```

**Parte 3 — Análisis con Pandas**📌




```python
#Añado nueva columna llamada total: df['total'] = df['precio'] * df['cantidad']

df1["total"] = df1["precio"] * df1["cantidad"]
```

```python
#Promedio de ventas
Promedio_ventas = df1['total'].mean()
print(Promedio_ventas)
```

```python
#Venta minima
Venta_minima = df1['total'].min()
print(Venta_minima)
```

```python
#Productos con mayor ventas
Productos_mayor_ventas = df1.groupby('producto')['total'].sum().sort_values(ascending=False)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
print(Productos_mayor_ventas)
```

```python
#Paises con más ventas
Pais_ventas = df1.groupby('pais')['total'].sum().sort_values(ascending=False)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
print(Pais_ventas)
```
