```
import re # Importo el modulo re, ya que me permitira remplazar valores especiales de texto de manera sencilla
```

```
df1 ["Name"] = df1["Name"].str.replace(r'[^\w\s]', '', regex=True) # Remplazo valores incoherentes
```

```
df1 ["2019"] = df1["2019"].replace("-",0)
```

```
df1 ["ChargesSheetingRate"] = df1["ChargesSheetingRate"].replace("-",0)
```

```
df1['State/UT'] = df1['State/UT'].replace({'UT': 'Ut'}) # Realizo remplazo de mayusculas para que mi base este siempre en un mismo formato
```

```
# Convierto a decimales

df1["RateofMurder"] = df1["RateofMurder"].astype(float)
df1["midPopulation"] = df1["midPopulation"].astype(float)
df1["ChargesSheetingRate"] = df1["ChargesSheetingRate"].astype(float)
```

```
# Convierto a enteros

df1["2019"] = df1["2019"].astype(int)
df1["2020"] = df1["2020"].astype(int)
df1["2021"] = df1["2020"].astype(int)
```

**Filtrado y selección de datos relevantes** 📌

La mayoria de las columnas más adelante me permitira hacer analisis más exactos y detallados sobre los homicidios sucedidos en la india entre el año 2019 y 2021, por ende, no realizo filtrado, ni selección.

**Transformación De Datos y creación de variables nuevas** 📌

```
df1["Total_homicides"]  = df1[['2019','2020','2021']].to_numpy().sum(axis=1) # Total de homicidios
```

```
# Nivel de peligro por estado

condiciones = [
    df1['Total_homicides'] < 500,
    (df1['Total_homicides'] >= 500)&(df1['Total_homicides'] < 1000),
    (df1['Total_homicides'] >= 100)&(df1['Total_homicides'] < 2000),
    df1['Total_homicides'] >= 2000
]

categorias = ['Low', 'Medium', 'High', 'Very high']

df1['Danger_level'] = np.select(condiciones, categorias, default='')
```

```
#Efectividad de la justicia en los casos denunciados

condicionescargos = [
    df1['ChargesSheetingRate'] <= 10.0,
    (df1['ChargesSheetingRate'] >= 10.0)&(df1['ChargesSheetingRate'] <= 40.0),
    (df1['ChargesSheetingRate'] >= 40.0)&(df1['ChargesSheetingRate'] <= 80.0),
    df1['ChargesSheetingRate'] >= 80.0
]

Nivel = ['✰', '⭐ ⭐', '⭐ ⭐', '⭐ ⭐ ⭐']

df1['effectiveness_justice'] = np.select(condicionescargos, Nivel, default='')
```

**Agrupación y resúmenes (Groupby)** 📌

```
# Estados con mas homicidios

States_from_highest_to_lowest_homicide_rates = df1.groupby("Name")["Total_homicides"].sum().sort_values(ascending=False).head(5)
print(States_from_highest_to_lowest_homicide_rates)
```

```
# Total casos en el estado y territorio de la unión

Cases_in_the_state_and_in_the_Union_Territory = df1.groupby("State/UT")["Total_homicides"].sum()
print(Cases_in_the_state_and_in_the_Union_Territory)
```

```
# Estados con mayor tasa de homicidios al transcurso de los años

Rates_of_murder = df1.groupby("Name")["RateofMurder"].mean().sort_values(ascending=False).head(5)
print(Rates_of_murder)
```

```
# Estados con menor tasa de homicidios al transcurso de los años

Rates_of_murder_low = df1.groupby("Name")["RateofMurder"].mean().sort_values(ascending=True).head(5)
print(Rates_of_murder_low)
```

```
# Tasa de población

Population_rate = df1.groupby("Name")["midPopulation"].mean().sort_values(ascending=False)
print(Population_rate)
```

```
average_total_cases = df1.groupby(['Name', 'midPopulation'], as_index=False)['Total_homicides'].mean().sort_values(by='Total_homicides', ascending=False)
average_total_cases.rename(columns={'Total_homicides': 'Average_Homicides'}, inplace=True)
print(average_total_cases)
```

```
# Estados con mayor tasa de justicia

Rates_of_justice = df1.groupby("Name")["ChargesSheetingRate"].mean().sort_values(ascending=False)
print(Rates_of_justice)
```

**¿Los estados con más homicidios tienden a tener tasas de cargos más altas o más bajas?** 📈



```
df1_sorted = df1.sort_values('Total_homicides', ascending=False)

posiciones = list(range(len(df1_sorted)))
ancho = 0.4

fig, ax = plt.subplots(figsize=(17, 6))

posiciones_izq = [i - ancho/2 for i in posiciones]
ax.bar(posiciones_izq, df1_sorted['Total_homicides'], width=ancho,
       label="Total homicidios", color="purple", alpha=0.85)
ax.set_ylabel("Total homicidios", color="purple")
ax.tick_params(axis="y", labelcolor="purple")

ax2 = ax.twinx()

posiciones_der = [i + ancho/2 for i in posiciones]
ax2.bar(posiciones_der, df1_sorted['ChargesSheetingRate'], width=ancho,
        label="Tasa de cargos (%)", color="deepskyblue", alpha=0.85)
ax2.set_ylabel("Tasa de cargos (%)", color="deepskyblue")
ax2.set_ylim(0, 120)
ax2.tick_params(axis="y", labelcolor="deepskyblue")

ax.set_xticks(posiciones)
ax.set_xticklabels(df1_sorted['Name'], rotation=45, ha="right", fontsize=8)

ax.set_title("Homicidios vs. tasa de cargos por estado", fontsize=14, fontweight="bold")

etiquetas1, nombres1 = ax.get_legend_handles_labels()
etiquetas2, nombres2 = ax2.get_legend_handles_labels()
ax.legend(etiquetas1 + etiquetas2, nombres1 + nombres2, loc="upper right")

plt.tight_layout()
plt.savefig("barras_comparativas.png", dpi=150, bbox_inches="tight")
plt.show()
```

Es posible afirmar que los estados que más tienen casos de homicidios tienen tasas de cargos muy de acuerdo a lo que deberian y efectivas, incluso podriamos decir que la justicia en estos años era bastante buena para los ciudadanos indios.

**¿Qué estados tuvieron mayores tasas de homicidios?** 📈

```
df1_sorted_homicides = df1.sort_values('Total_homicides', ascending=False)

plt.figure(figsize=(17, 6))
bars = plt.bar(df1_sorted_homicides['Name'], df1_sorted_homicides['Total_homicides'], color='purple')

for bar in bars:
    altura = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, altura + 50,
             str(int(altura)), ha='center', va='bottom', fontsize=7)

plt.xlabel('Estados')
plt.ylabel('Total de Homicidios')
plt.title('Total homicidios por Estado', fontsize=14, fontweight='bold')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

El top 5 de estados que mayor tasa de homicidios tuvieron fue:

1. Ulttar Pradesh: 11.364 homicidios
2. Bihar: 9.438 homicidios
3. Maharashtra: 6.468 homicidios
4. Madhya Pradesh: 5.997 homicidios
5. West Bengal: 5.826 homicidios

Esto se puede deber a que en Uttar Pradesh y Bihar ha enfrentado multiples desafios relacionados a la desigualdad economica, el desempleo y conflictos. Mientras que Maharashtra y West Bengal cuentan con grandes áreas metropolitanas donde la concentración de población pudo favorecer una mayor tasa de homicidios en estos años. Por su parte, Madhya Pradesh siempre ha presentado niveles elevados de peligrosidad en comparación con otros estados.

**¿Cómo ha cambiado el número de homicidios a lo largo de los años en la india?** 📈

```
homicidios_por_año = {
    '2019': df1['2019'].sum(),
    '2020': df1['2020'].sum(),
    '2021': df1['2021'].sum()
}

años = list(homicidios_por_año.keys())
valores = list(homicidios_por_año.values())

plt.figure(figsize=(7, 7))
colores = ['purple', 'deepskyblue', 'mediumpurple']

wedges, texts, autotexts = plt.pie(
    valores,
    labels=años,
    colors=colores,
    autopct='%1.1f%%',
    startangle=140
)

for text in autotexts:
    text.set_fontsize(11)
    text.set_fontweight('bold')

plt.title('Homicidios a lo largo de los años', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

Se evidencia un ligero incremento entre el año 2019 y 2020, pasando de 33.1% (28.915) a 33.4% (29.193) y mateniendo esta misma cantidad para el año 2021. Por ende, podriamos determinar que la tendencia es estable y los factores que influyen en el aumento de casos sigue estando presente.

**¿Se registraron más casos de homicidios en los estados que en las zonas federales (territorio de la unión)?** 📈

```
states = df1[df1['State/UT'] == 'State']
uts = df1[df1['State/UT'] == 'Ut']

conteo = {
    'States': len(states),
    'Union Territories': len(uts)
}

etiquetas = list(conteo.keys())
valores = list(conteo.values())

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(17, 5))

bars = ax1.bar(etiquetas, valores, color=['purple', 'deepskyblue'], alpha=0.85, width=0.5)

for bar in bars:
    altura = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, altura + 0.1,
             str(int(altura)), ha='center', va='bottom', fontsize=12, fontweight='bold')

ax1.set_title('State vs Union Territories', fontsize=13, fontweight='bold')
ax1.set_ylabel('Cantidad')
ax1.set_ylim(0, max(valores) + 3)

wedges, texts, autotexts = ax2.pie(
    valores,
    labels=etiquetas,
    colors=['purple', 'deepskyblue'],
    autopct='%1.1f%%',
    startangle=140
)

for text in autotexts:
    text.set_fontsize(11)
    text.set_fontweight('bold')

ax2.set_title('Distribución States vs UT', fontsize=13, fontweight='bold')
plt.suptitle('Clasificación por tipo de región', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

Es correcto afirmar que el 77.8% de los estados, la gran mayoria pertenecen al states y solo el 22.2 a zonas federales. Esto puede estar relacionado con la tasa de homicidios, ya que en zonas más apartadas y menos pobladas la criminalidad es menor, por ende, menos casos y menos peligro.
